### Question 1

#### Objectif

Comparer le risque moyen de l’estimateur OLS  
$$
\mathbb{E}[\,R_X(\hat\theta)\,] \;=\;\frac{n-d}{n}\,\sigma^2
$$  
au risque bayésien (oracle connaissant $\theta^*$)  
$$
R_{\text{Bayes}} \;=\;\sigma^2.
$$

---

#### Bayes risk

Dans le modèle linéaire  
$$
y \;=\; X\,\theta^* + \varepsilon,\quad \varepsilon\sim\mathcal{N}(0,\sigma^2I_n),
$$  
si on connaît $\theta^*$ alors la prédiction est $\hat y = X\theta^*$ et  
$$
R_{\text{Bayes}}
=\mathbb{E}\!\Bigl[\tfrac1n\|y - X\theta^*\|^2\Bigr]
=\mathbb{E}\!\Bigl[\tfrac1n\|\varepsilon\|^2\Bigr]
=\sigma^2.
$$

---

#### Risque de l’estimateur OLS

Quand on estime $\hat\theta=(X^TX)^{-1}X^Ty$ sur les mêmes données $X$, on obtient  
$$
\mathbb{E}[\,R_X(\hat\theta)\,]
=\frac{n - d}{n}\,\sigma^2.
$$  
Comme $\tfrac{n-d}{n}<1$ pour $d>0$, on a  
$$
\mathbb{E}[\,R_X(\hat\theta)\,]\;<\;R_{\text{Bayes}}.
$$

- **Risque OLS entraînement** : plus faible car on applique $\hat\theta$ sur **les mêmes** $X$ que ceux utilisés pour estimer $\theta^*$ (léger sur-apprentissage / « optimisme » du risque entraînement).

### Question 3

$$
\sum_{(i,j) \in [1,n]^2} A_{ij}^2 = \operatorname{tr}(A^T A)
$$


On considère une matrice $ A \in \mathbb{R}^{n \times n} $. On souhaite montrer que la somme des carrés des coefficients de $ A $ est égale à la trace de $ A^T A $.

1. Par définition du produit matriciel, l’entrée $ (k, k) $ de $ A^T A $ s’écrit :

$$
(A^T A)_{kk} = \sum_{l=1}^n A_{lk} A_{lk} = \sum_{l=1}^n A_{lk}^2
$$

2. La trace d'une matrice est la somme de ses éléments diagonaux, donc :

$$
\operatorname{tr}(A^T A) = \sum_{k=1}^n (A^T A)_{kk} = \sum_{k=1}^n \sum_{l=1}^n A_{lk}^2
$$

3. En permutant les indices $ i = l $, $ j = k $, on obtient :

$$
\operatorname{tr}(A^T A) = \sum_{i=1}^n \sum_{j=1}^n A_{ij}^2 = \sum_{(i,j)\in [1,n]^2} A_{ij}^2
$$


### Question 4

$$
\|A \varepsilon\|^2 = (A \varepsilon)^T (A \varepsilon) = \varepsilon^T A^T A \varepsilon
$$

On prend ensuite l'espérance :

$$
\mathbb{E} \left[ \frac{1}{n} \| A \varepsilon \|^2 \right] = \frac{1}{n} \mathbb{E} \left[ \varepsilon^T A^T A \varepsilon \right]
$$

Or, si $\varepsilon \sim \mathcal{N}(0, \sigma^2 I_n)$, on a la formule classique :

$$
\mathbb{E} \left[ \varepsilon^T M \varepsilon \right] = \sigma^2 \operatorname{tr}(M)
\quad \text{(pour toute matrice symétrique $M$)}
$$

Ici, $M = A^T A$, donc :

$$
\mathbb{E} \left[ \varepsilon^T A^T A \varepsilon \right] = \sigma^2 \operatorname{tr}(A^T A)
$$

On conclut donc :

$$
\mathbb{E}_\varepsilon \left[ \frac{1}{n} \| A \varepsilon \|^2 \right] = \frac{\sigma^2}{n} \operatorname{tr}(A^T A)
\quad
$$


### Question 5

On nous donne :

$$
P := X (X^T X)^{-1} X^T
$$

Alors :

$$
A = I_n - P
$$

#### 2. Propriétés de $P$ :

- $P^T = P$ (symétrique)
- $P^2 = P$ (idempotente)

Donc $P$ est une **matrice de projection orthogonale** sur l’image de $X$.

#### 3. Symétrie de $A$ :

$$
A^T = (I_n - P)^T = I_n^T - P^T = I_n - P = A
$$

Donc $A$ est **symétrique**.

#### 4. Calcul de $A^2$ :

$$
A^2 = (I_n - P)^2 = I_n - 2P + P^2
$$

Or $P^2 = P$, donc :

$$
A^2 = I_n - 2P + P = I_n - P = A
$$

Donc $A$ est **idempotente**.



### Conclusion :

Puisque $A$ est symétrique et idempotente :

$$
A^T A = A
$$


### Question 7

On cherche à calculer :

$$
\mathbb{E} \left[ \frac{ \| y - X\hat\theta \|^2 }{ n - d } \right]
$$

Or, on a démontré à la question précédente que :

$$
\mathbb{E} \left[ R_X(\hat\theta) \right] = \mathbb{E} \left[ \frac{1}{n} \| y - X\hat\theta \|^2 \right] = \frac{n - d}{n} \sigma^2
$$

Ce qui implique que :

$$
\mathbb{E} \left[ \| y - X\hat\theta \|^2 \right] = (n - d) \sigma^2
$$

Donc :

$$
\mathbb{E} \left[ \frac{ \| y - X\hat\theta \|^2 }{ n - d } \right] = \sigma^2
$$

```python
import numpy as np

n = 100
d = 10
sigma2 = 1.0
theta_star = np.random.randn(d)

X = np.random.randn(n, d)

epsilon = np.random.randn(n) * np.sqrt(sigma2)

y = X @ theta_star + epsilon

theta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

residuals = y - X @ theta_hat

sigma2_hat = (np.linalg.norm(residuals)**2) / (n - d)

print(f"Valeur théorique de σ² : {sigma2}")
print(f"Estimation de σ² par la formule : {sigma2_hat:.4f}")
```
